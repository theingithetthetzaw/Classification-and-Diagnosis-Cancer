```python
X=pd.read_csv("breast-cancer.csv",header=None,usecols=range(0,9),skiprows=1).dropna()
X
```

```python
Y=pd.read_csv("breast-cancer.csv",header=None,usecols=range(9,10),skiprows=1).dropna()
Y
```

```python
X.shape
#Y.shape
```

```python
X=np.array(X)
Y=np.array(Y)
```

```python
from sklearn import preprocessing
```

```python
le=preprocessing.LabelEncoder()
le.fit(Y)
Y=le.transform(Y)

print(Y)
```

```python
for j in range(0,9):
        le.fit(X[:,j])
        X[:,j]=le.transform(X[:,j])
X
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=12)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
```

```python
y_pred_train=gnb.predict(X_train)
print(y_pred_train)
```

```python
print((Y_train!=y_pred_train).sum())
```

```python
y_pred_test=gnb.predict(X_test)
print(y_pred_test)
```

```python
print((Y_test!=y_pred_test).sum())
```

```python
from sklearn.metrics import accuracy_score
```

```python
tr_acc=accuracy_score(Y_train, y_pred_train)
print("Train accuracy %s" %(tr_acc*100 ))
```

```python
te_acc=accuracy_score(Y_test, y_pred_test)
print("Train accuracy %s" %(te_acc * 100))
```
